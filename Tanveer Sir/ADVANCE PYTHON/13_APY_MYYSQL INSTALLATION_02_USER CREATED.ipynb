NOTE:
TCL operations perform only on DML queries.

DELETE: 

It is used to delete one row, specified rows and all rows of table but it will not delete structure of the table,we can rollback after delete.

TRUNCATE:

It is used to delete all rows of table but it won't delete structure of the table.We cant rollback after truncate and we cant use 'where' with truncate.

DROP:

It is used to delte data as well as structure of the table.We cant rollback after drop.
