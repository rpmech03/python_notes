# METHODS OF STRING :

1_CAPITALIZE :Mainly used to convert "first letter" of string in "uppercase" and rest will be in lowercase.

2_CASEFOLD : Used to convert in lowercase when we are compairing two string.

3_CENTER : used to return a centered align string.

4_COUNT : To count occurence of substrings count in string.

5_ENDSWITH :Used to check a string if ending with a specific substring.

6_EXPAND TABS : To pass tab size of the string.

7_FIND : Used to search a specific substring with in a string and return lowest index of substring.

8_FORMAT : To format specific value of string.

9_FORMAT-MAP : little bit use

10_INDEX : Used to search substring within string and return lowest index value when substring will find otherwise raise a 

           value error,exceptions.
           
11_IS ALANUM : It returns true when string contains alphabets.

12_JOIN : Used to join elements of a collection , object like (list,tuple,set,dict) to the end of string.

13_L-JUST : To return list justified string.

14_LOWER : Used to convert string into lowercase.

15_L-STRIP : To remove space from left side of a string.

16_MAKETRANS : Used for string translation.

17_PARTITION : To divide a string in 3 parts and return as a tuple.

18_REPLACE : To replace some content of string with another content.

19_Z FILL :

20_RFIND : To search element from highest index and return highest index when element will found.

21_RINDEX : To search a specific element from highest index in string and return specific index when element will find       

            otherwise produce error.
            
22_RJUST : Toreturn right justified string.

23_RPARTITIONS : It is same as partition but it will make partition from right side.

24_RSPLIT : To split a string from right side according to substring.

25_RSTRIP : Remove space from right side of a string.

26_SPLIT : To split a string according to a substring and return list of substring.

27_STARTWITH : Return a true when a string start with specified substring.
